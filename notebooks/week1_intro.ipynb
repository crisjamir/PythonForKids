## 3 — Variables and basic data types

A *variable* is a name that refers to a value. Python has several basic data types: integers (`int`), floating point numbers (`float`), strings (`str`), and booleans (`bool`).


```python
# Examples of variables and types
a = 10           # int
b = 3.14         # float
name = 'Alice'   # str
is_student = True  # bool

print('a =', a, 'type:', type(a))
print('b =', b, 'type:', type(b))
print("name =", name, 'type:', type(name))
print('is_student =', is_student, 'type:', type(is_student))
```

## 4 — Basic arithmetic and expressions

Python can be used as a calculator. Try the examples below and add your own.


```python
x = 7
y = 3
print('x + y =', x + y)
print('x - y =', x - y)
print('x * y =', x * y)
print('x / y =', x / y)   # float division
print('x // y =', x // y) # integer (floor) division
print('x % y =', x % y)   # remainder
print('x ** y =', x ** y) # power
```
