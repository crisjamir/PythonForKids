# Week 1 — Exercises (Set of 10)

This notebook contains **10 practice exercises** focused on basic input/output, type conversion, basic operations, variables, and data types. For each exercise replace `# TODO` with your solution and run the cell.

**How to submit:** Save this notebook as `notebooks/week1_exercises_set.ipynb` and upload/commit it to your repository.

> Estimated time: 2–3 hours


## Exercise 1 — Hello and age
Write a program that asks for the user's name and age (age as integer). Print a greeting: `Hello <name>, you are <age> years old.`

## Exercise 2 — Swap values
Read two values (could be numbers or strings) from the user and print them swapped. Do not use a third variable (use tuple unpacking).

## Exercise 3 — Celsius to Fahrenheit
Ask the user for temperature in Celsius (can be float). Convert to Fahrenheit using `F = C * 9/5 + 32` and print with one decimal place.

## Exercise 4 — Area of a rectangle
Ask for width and height (floats). Compute area and print `Area = <value>`.

## Exercise 5 — Integer division and remainder
Ask the user for two integers `a` and `b`. Print the quotient (integer division) and remainder. Handle division by zero gracefully.

## Exercise 6 — Type detective
Read a value from input and try to detect whether it is an `int`, `float`, or `str`. Print the detected type. (Hint: attempt conversions in order.)

## Exercise 7 — Average of three numbers
Ask for three numbers (floats). Compute and print their average with 2 decimal places.

## Exercise 8 — String formatting
Ask for first name and last name. Print them in the format `Last, First` and also print the total number of characters (excluding spaces).

## Exercise 9 — Simple tip calculator
Ask for bill amount (float) and tip percentage (int). Compute tip and total amount. Print both with 2 decimal places.

## Exercise 10 — Simple data entry to CSV
Ask the user for name, age (int), and email. Append this as a CSV line to `data/contacts.csv`. Create the `data/` folder if it doesn't exist. (Hint: use the `os` module.)
